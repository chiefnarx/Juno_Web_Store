### Data Extraction

```python
url = 'https://www.jumia.com.ng/catalog/?q=laptops'

r = requests.get(url)

soup = BeautifulSoup(r.text, "lxml")
```

```python
# Product names
product_names = []
names = soup.find_all("h3", class_ = "name")
for name in names:
    product_names.append(name.string)
```

```python
product_names
```

```python
# Current price
current_prices = []
prices = soup.find_all("div", class_ = "prc")
for price in prices:
    current_prices.append(price.string)
```

```python
current_prices
```

```python
# Former price
former_prices = []
prices = soup.find_all("div", class_ = "old")
for price in prices:
    former_prices.append(price.string)
```

```python
former_prices
```

```python
# Discount percentages
discount_percs = []
discounts = soup.find_all("span", class_ = "bdg_dsct _dyn -mls")
for discount in discounts:
    discount_percs.append(discount.string)
```

```python
discount_percs
```

```python
data = {
    'title': product_names, 
    'current_price': current_prices, 
    'former_prices': former_prices, 
    'discount_percs': discount_percs
    }

laptop_df = pd.DataFrame.from_dict(data, orient='index')
```

```python
laptop_df = laptop_df.transpose()
laptop_df
```

```python
pages = [x for x in range(1,51)]

product_names = []
current_prices = []
former_prices = []
discount_percs = []

for page in pages:
    url = 'https://www.jumia.com.ng/catalog/?q=laptops&{page}=50#catalog-listing'
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml")

    # Product names
    names = soup.find_all("h3", class_ = "name")
    for name in names:
        product_names.append(name.string)

    # Current price
    prices = soup.find_all("div", class_ = "prc")
    for price in prices:
        current_prices.append(price.string)

    # Former price
    prices = soup.find_all("div", class_ = "old")
    for price in prices:
        former_prices.append(price.string)

    # Discount percentage
    discounts = soup.find_all("span", class_ = "bdg_dsct _dyn -mls")
    for discount in discounts:
        discount_percs.append(discount.string)

data = {
    'title': product_names, 
    'current_price': current_prices, 
    'former_prices': former_prices, 
    'discount_percs': discount_percs
    }
laptop_df = pd.DataFrame.from_dict(data, orient='index')
laptop_df = laptop_df.transpose()
laptop_df
```

```python
laptop_df.info()
```

```python
laptop_df.fillna({
    'former_price': '0',
    'discount_percs': '0%'
}, inplace=True)

laptop_df
```
